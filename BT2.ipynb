Cài đặt min edit distance




```python
source = "intention"
target = "execution"

n = len(source)
m = len(target)

def sub_cost(char1, char2):
    if char1 == char2:
        return 0
    if char1 != char2:
        return 2

def min_edit_distance(source, target):
    n = len(source)
    m = len(target)
    D = [[0]*(m+1) for _ in range(n+1)]

    D[0][0] = 0
    for i in range(1,len(D)):
        D[i][0] = D[i-1][0] + 1
    for j in range(1,len(D[0])):
        D[0][j] = D[0][j-1] + 1

    for i in range(len(source)):
        for j in range(len(target)):
            D[i+1][j+1] = min(D[i][j+1] + 1, \
                          D[i][j] + sub_cost(source[i],target[j]), \
                          D[i+1][j] + 1)
    # print(D)
    return D[-1][-1]



```

```python
min_edit_distance("intention", "execution")
```

```python
def min_edit_distance_backtrace(source, target):
    n = len(source)
    m = len(target)
    D = [[0]*(m+1) for _ in range(n+1)]
    ptr = [[[] for i in range(m+1)] for j in range(n+1)]
    D[0][0] = 0


    for i in range(1,len(D)):
        D[i][0] = D[i-1][0] + 1
        ptr[i][0].append(f"Delete {source[i-1]}")
    for j in range(1,len(D[0])):
        D[0][j] = D[0][j-1] + 1
        ptr[0][j].append(f"Insert {target[j-1]}")
    # print(ptr)

    for i in range(1,len(source)+1):
        for j in range(1,len(target)+1):
            D[i][j] = min(D[i-1][j] + 1, \
                          D[i-1][j-1] + sub_cost(source[i-1],target[j-1]), \
                          D[i][j-1] + 1)
            if D[i-1][j] + 1 == D[i][j]:
                ptr[i][j] = ptr[i-1][j] + ["Delete {}".format(source[i-1])]
            if D[i][j-1] + 1 == D[i][j]:
                ptr[i][j] = ptr[i][j-1] + ["Insert {}".format(target[j-1])]
            if D[i-1][j-1] + 2 == D[i][j]:# if source[i-1] != target[j-1]:
                ptr[i][j] = ptr[i-1][j-1] + ["Substitute {0} with {1}".format(source[i-1],target[j-1])]
            else:
                ptr[i][j] = ptr[i-1][j-1]

    return D, ptr
```

Needleman - Wunsch


```python
dna1 = "GCATGCG"
dna2 = "GATTACA"

n = len(source)
m = len(target)

def score(letter1, letter2):
    if letter1 == letter2: ##match
        return 1
    else: #mismatch or indel (insertion or deletion)
        return -1

def needleman_wunsch(DNA1, DNA2):
    n = len(DNA1)
    m = len(DNA2)
    d = -1 #gap ?? (insert or delete)
    D = [[0]*(m+1) for _ in range(n+1)]

    D[0][0] = 0
    for i in range(1,len(D)):
        D[i][0] = i*d
    for j in range(1,len(D[0])):
        D[0][j] = j*d

    for i in range(1, n+1):
        for j in range(1, m+1):
            D[i][j] = max(D[i-1][j] + d, #delete
                          D[i-1][j-1] + score(DNA1[i-1],DNA2[j-1]), #match/mismatch
                          D[i][j-1] + d) #insert
    # print(D)
    return D



```

```python
needleman_wunsch(dna1,dna2)
```
